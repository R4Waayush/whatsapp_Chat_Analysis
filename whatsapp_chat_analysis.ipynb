```python
import pandas as pd
import re
```

```python
f = open('WhatsApp Chat with Stake.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s(?:AM|PM)?\s?-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
 dates = re.findall(pattern, data)
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
```

```python
dates
```

```python
df['message_date'] = df['message_date'].str.replace(' -', '', regex=False).str.strip()
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %I:%M %p')
```

```python
df.head()
```

```python

users = []
messages = []
for message in df['user_message']:    # <-- if 'user_message' still exists
    entry = re.split(r'([\w\W]+?):\s', message, maxsplit=1)
    if len(entry) > 2:  # if sender name and message exist
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages

df.drop(columns=['user_message'], inplace=True)   # drop only after splitting

df.head();
```

```python
df.head();
```

```python
df.rename(columns={'message_date': 'date'}, inplace=True)
```

```python
df.head()
```

```python
df['month'] = df['date'].dt.month_name()
```

```python
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python
df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
df['month_num'] = df['date'].dt.month
```
